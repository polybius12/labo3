```python
def entrenar_arima(data):
  import pandas as pd
  import numpy as np
  from statsmodels.tsa.arima.model import ARIMA
  from sklearn.metrics import mean_absolute_error, mean_squared_error


  # Función de preprocesamiento
  def preprocess_data(data):
  #    data['periodo'] = pd.to_datetime(data['periodo'])
  #   data.set_index('periodo', inplace=True)
      return data

  # Función para la división de datos en entrenamiento, validación y predicción
  def split_data(data, train_months, val_months, pred_months):
      n = len(data)
      total_months = train_months + val_months + pred_months
      if total_months > n:
          return 0, 0, 0

      else:
        train_data = data.iloc[:train_months]
        val_data = data.iloc[train_months:train_months + val_months]
        pred_data = data.iloc[train_months + val_months:train_months + val_months + pred_months]

        return train_data, val_data, pred_data

  # Función para calcular métricas de error
  def calculate_metrics(actual, predicted):
      mae = mean_absolute_error(actual, predicted)
      mse = mean_squared_error(actual, predicted)
      rmse = np.sqrt(mse)
      mape = np.mean(np.abs((actual - predicted) / actual)) * 100
      return mae, mse, rmse, mape

  # Función para modelar con ARIMA
  def arima_forecast(train_data, val_data, order=(1, 1, 1)):
      model = ARIMA(train_data, order=order)
      model_fit = model.fit()
      forecast = model_fit.forecast(steps=len(val_data))
      return forecast

  # Función principal
  def main(data, train_months, val_months, pred_months, order=(1, 1, 1)):
      data = preprocess_data(data)
      train_data, val_data, pred_data = split_data(data['tn'], train_months, val_months, pred_months)

      # Validación hacia adelante
      predictions = []
      for i in range(len(val_data)):
          current_train = pd.concat([train_data, val_data[:i]])
          current_val = val_data[i:i+1]
          forecast = arima_forecast(current_train, current_val, order)
          predictions.append(forecast.iloc[0])

      mae, mse, rmse, mape = calculate_metrics(val_data, predictions)
      print(f"Validation Metrics: MAE={mae}, MSE={mse}, RMSE={rmse}, MAPE={mape}%")

      # Predicción futura
      future_forecast = arima_forecast(pd.concat([train_data, val_data]), pred_data, order)
      return future_forecast.values[0]

  return main(data, train_months=12, val_months=1, pred_months=1, order=(1, 1, 1))

```
